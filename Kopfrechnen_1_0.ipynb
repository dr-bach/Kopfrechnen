##Kopfrechnen Version 1.0
- mehrere SpielerInnen (Identifizierung über Namen)
- Spielstand speichern
- neue Datenstruktur für richtige bzw. falsche Antworten


Indem wir den Spielstand laden und speichern, ermöglichen wir die Erzeugung von Daten, aus denen sich Informationen entnehmen lassen. Dies ist generell die Grundlage für Data Science und wird uns in späteren Erweiterungen nützen. Daher nutzen wir hier erneut DataFrames - das Modul "pandas" importieren wir daher schon bei den Initialisierungen.

```python
#@title Code Kopfrechnen 1.0

#initialisierungen

import random #für Zufallszahlen

import os.path #für Dateizugriff

import pandas as pd #für Datenstruktur

#inhalt helper
def addition(min, max):
  while True:
    (a,b)=zufallszahlen(min,max)
    if (a+b)<=max:
      break
  return (a,b,a+b)

def subtraktion(min, max):
  while True:
    (a,b)=zufallszahlen(min,max)
    if (a-b)>=min:
      break
  return (a,b,a-b)

def multiplikation(min, max):
  (a,b)=zufallszahlen(min,max)
  return (a,b,a*b)

def division(min, max):
  while True:
    (a,b)=zufallszahlen(min,max)
    dividend=a*b
    if (a*b)<=max*max:
      break
  return (dividend,b,a)

def chart_pie(ergebnis):
  ergebnis.plot.pie(y='Anzahl', labels=ergebnis['Bewertung'], colors = ('lightgreen','red'),
              autopct='%.0f%%', startangle=60)


#leeren Datensatz initialisieren
data = {'Bewertung': ['richtig','falsch'],
        'Anzahl': [0,0]}

def laden(datei):
  if not os.path.isfile(datei):
    auswertung = pd.DataFrame(data)
    auswertung.to_csv (datei,index=False,header=True)
  auswertung = pd.read_csv(datei,index_col=False)
  return auswertung

def identifizieren():
  id=input("Bitte ID eingeben:")
  if id=="":
    datei="kopfrechnen.csv"
  else:
    datei=id+".csv"
  return datei

def speichern(datei, auswertung):
  auswertung.to_csv (datei,index=False,header=True)


#aufgabe erstellen

def zufallszahlen(min, max):
  return(random.randint(min, max),random.randint(min, max))

def aufgabe_stellen():
  rechenart=random.choice(["+","-","*",":"])
  if rechenart == "+":
    (a,b,ergebnis) = addition(1,100)
  elif rechenart == "-":
    (a,b,ergebnis) = subtraktion(1,100)
  elif rechenart == "*":
    (a,b,ergebnis) = multiplikation(1,10)
  else:  #rechenart == ":"
    (a,b,ergebnis) = division(1,10)
  print ("wieviel ist {0} {1} {2}?".format(a, rechenart, b))
  return ergebnis


#eingabe bewerten
  
def abfrage():
  while True:
    try:
      text = input("Dein Ergebnis = ")
      if text=="":
        return -1
      eingabe = int(text)
      break
    except ValueError:
      print("Bitte eine Zahl eingeben")     
  return eingabe

def bewerten (ergebnis, eingabe):
  if ergebnis==eingabe:
    print("Richtig\n")
    return 1
  elif ergebnis<0:
    return -1
  else:
    print(eingabe,"ist falsch. Das richtige Ergebnis lautet", ergebnis,"\n")
    return 0


#hauptprogramm

def spielen():
  print("Willkommen beim Kopfrechnen.")
  datei=identifizieren()
  auswertung = laden(datei)
  richtig=auswertung.Anzahl[0] #entspricht der 1. Zeile des DataFrame
  falsch=auswertung.Anzahl[1] #entspricht der 2. Zeile des DataFrame
  runden=richtig + falsch
  print ("Du hast bisher {0} von {1} Aufgaben richtig gerechnet".
         format(richtig, runden))
  alte_quote=(richtig/runden*100).round(2)
  print("Willkommen beim Kopfrechnen. Dir werden verschiedene Aufgaben gestellt.")
  print("Wenn Du das Spiel beenden willst, drücke statt der Antwort nur die ENTER-Taste.")
  while True:
      ergebnis=aufgabe_stellen()
      eingabe=abfrage()
      if eingabe<0:
        break
      else: 
        richtig += bewerten (ergebnis, eingabe) # +1 wenn richtig, +0 wenn falsch
      runden += 1
  print ("Du hast insgesamt {0} von {1} Aufgaben richtig".
         format(richtig, runden))
  quote=(richtig/runden*100).round(2)
  print ("Das entspricht einer Quote von",quote,"%")
  if quote>alte_quote:
    print("Prima, Du hast dich verbessert")
  auswertung.Anzahl[0]=richtig
  auswertung.Anzahl[1]=runden-richtig
  speichern(datei, auswertung)
  chart_pie(auswertung)

spielen()
```
